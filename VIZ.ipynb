```python
plt.style.use('ggplot')
c = []

for value in tip['sex'].values:
    if value == 'Male':
        c.append('red')

    else:
        c.append('blue')
        
```

```python
plt.plot(dpi=100)
plt.scatter(tip['total_bill'],tip['tip'],c=c)


## For the markinf scheme we will have tje follwing

plt.plot([],[],'ro',label='Male')

plt.plot([],[],'ko',label='Female',color = 'gray')




plt.legend()
plt.grid()

plt.show()
```
